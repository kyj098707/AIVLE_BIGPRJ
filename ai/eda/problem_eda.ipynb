### 함수 호출

```python
# 문자열을 리스트로 변환
def str_to_list(x):
    try:
        return literal_eval(x)
    except: #해당 값이 null값이거나 오류가 있을 때, None을 return 하기
        return None
    
# 딕셔너리에서 키값만 반환
def dic_to_list(x):
    try:
        temp = []
        for i in x:
            temp.append(i["key"])
        if not temp:
            return None
        else:
            return temp
    except: #해당 값이 null값이거나 오류가 있을 때, None을 return 하기
        return None

def box_violin(x, lim):
    ax = sns.boxplot(x = x)
    ax = sns.violinplot(x =x)

    ax2 = plt.axes([0.6, 0.6, .2, .2])
    ax2 = sns.boxplot(x = x)
    ax2.set_xlim([0, lim])
    ax2.set_title('zoom')
```

###  코사인 유사도

```python
from sklearn.metrics.pairwise import cosine_similarity as cs
problem_idx = 953

#코사인 유사도
problem_test = problem[['problemId', 'titleKo', 'level', 'averageTries']]
set_tags = set(problem.tags.dropna().explode())
for i in list(set_tags):
    problem_test[i] = 0

for i in tqdm(range(len(problem))):
    tags = problem.iloc[i, 6]
    try:
        for tag in tags:
            problem_test.iloc[i, list(problem_test.columns).index(tag)] += 1
    except:
        pass
temp = problem_test.iloc[:, 2:].values
sim = cs(temp, temp)
sim_scores = list(enumerate(sim[problem_idx]))
sim_scores = sorted(sim_scores, key=lambda x: x[1], reverse=True)
sim_scores = sim_scores[1:20]
print(sim_scores)
```

```python
idx = np.array(sim_scores)[:, 0]
df_cos = problem_test.iloc[idx]
df_cos
```

### 유클리드 거리

```python
def dist(x,y):   
    return np.sqrt(np.sum((x-y)**2))
```

```python
list_dist = []
for i, j in tqdm(enumerate(temp)):
    list_dist.append((i, dist(temp[problem_idx], j)))
list_dist = sorted(list_dist, key = lambda x : x[1])
list_dist = list_dist[:20]
print(list_dist)
```

```python
df_euclidean = problem_test.iloc[np.array(list_dist)[1:, 0]]
df_euclidean
```

코사인 유사도로 구한 방식보다 레벨이 전반적으로 비교 대상인 2023번이랑 비슷함

###  자카드 유사도

```python
def jaccard(x, y):
    union = set(x).union(set(y))
    inter = set(x).intersection(set(y))
    return len(inter)/len(union)
```

```python
list_jaccard = []
for i, j in tqdm(enumerate(temp)):
    list_jaccard.append((i, jaccard(temp[problem_idx], j)))
list_jaccard = sorted(list_jaccard, key = lambda x : x[1], reverse=True)
list_jaccard = list_jaccard[:60]
print(list_jaccard)
list_jaccard
```

```python
df_jaccard = problem_test.iloc[np.array(list_jaccard)[1:, 0]]
df_jaccard.head(20)
```

자카드 유사도는 교집합, 합집합으로 유사한 문제를 찾기 떄문에 예시인 2023번 문제와 태그와 레벨이 비슷함

##### 3. 타겟 유저가 푼 태그를 기반으로 부족한 부분을 더 많이 푼 유저

```python
userproblem = pd.merge(user, solved_problem, on='handle')
# null이 담긴 사용자 제거
userproblem = userproblem[~userproblem.solved_problem.isnull()]
userproblem
```

```python
problem = problem[problem.level != 0]
# not_solvable == False만
problem = problem[problem.isSolvable == True]
# tag가 nan인 문제 제거
problem = problem[~problem.tags.isnull()]
problem.head()
```

```python
# 후보 유저들만 추리기
candi_users = candi_by_rbps.handle.values
candi_by_tags = userproblem[userproblem.handle.isin(candi_users)]
candi_by_tags
```

```python
# 필요한 태그 칼럼 정리하기
tags = ['math', 'implementation', 'greedy', 'string', 'data-structures', 'geometry', 'dp', 'dfs', 'bfs']
for x in tags:
    candi_by_tags[x] = 0

candi_by_tags
```

```python
def class_user_tag(df):    
    # 특정 유저가 푼 problems 구하기
    for user in tqdm(candi_users):
        if (len(df[df.handle == user].solved_problem.values) == 0):
            continue;
        df_one_user_problems = df[df.handle == user].solved_problem.values[0]
        one_user_problems = ','.join(df_one_user_problems).split(',')
        
        # 특정 유저가 푼 problems의 태그 값들 구하기
        user_tag = []
        # for i in range(len(one_user_problems)):
        #     one_problem = problem[problem.problemId == int(one_user_problems[i])]
        #     one_problem_tags = ','.join([','.join(tags) if isinstance(tags, list) else str(tags) for tags in one_problem.tags.values]).split(',')
        #     user_tag.extend(one_problem_tags)
        
        for i in range(len(one_user_problems)):
            problem_id = one_user_problems[i]
            if problem_id != '':
                one_problem = problem[problem.problemId == int(problem_id)]
                one_problem_tags = ','.join([','.join(tags) if isinstance(tags, list) else str(tags) for tags in one_problem.tags.values]).split(',')
                user_tag.extend(one_problem_tags)


        # 특정 유저가 푼 태그별 개수 구하기: e.g. counts = {'arithmetic': 1, 'implementation': 3, 'math': 10, ... }
        counts = dict()
        for i in user_tag:
            counts[i] = counts.get(i, 0) + 1

        # 특정 유저의 태그 칼럼에 각 태그 값 넣기
        idx = df[df['handle'] == user].index
        for k, v in counts.items():
            if k in df.columns:
                df.loc[idx, k] = v

    return df
```

```python
df_candi_with_tags = class_user_tag(candi_by_tags)
```

```python
# 라이벌 유저군들의 푼 태그 수와 타겟 유저가 푼 태그 수 확인
df_target_candi_tags = pd.DataFrame(df_candi_with_tags.sum(axis=0)[tags]).reset_index()
df_target_candi_tags.columns = ['tags', 'num']
df_target_candi_tags.num = df_target_candi_tags.num/len(df_candi_with_tags)

target_tags_values = df_candi_with_tags[df_candi_with_tags.handle == target_user][tags].values
df_target_candi_tags['target'] = target_tags_values[0]
df_target_candi_tags
```

```python
# 라이벌 유저군 보다 내가 부족한 태그 부분 확인
bar_width = 0.35
alpha = 0.5
index = np.arange(9)
label = df_target_candi_tags.tags

plt.figure(figsize=(15,5))
p1 = plt.bar(index, df_target_candi_tags.target, bar_width, 
             color='b', alpha=alpha,label=target_user)

p2 = plt.bar(index + bar_width, df_target_candi_tags.num, bar_width, 
            color='r', alpha=alpha,label='mean_users')

plt.title('Compare solved tags of '+target_user+' and mean of users in same user_class', fontsize=15)
plt.ylabel('tags #', fontsize=10)
#plt.xlabel('tags', fontsize=18)
plt.xticks(index, label, fontsize=15, rotation=45)
plt.legend((p1[0], p2[0]), (target_user, 'rival group users'), fontsize=15)
plt.show()
```
